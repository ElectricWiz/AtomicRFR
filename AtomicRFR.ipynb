```python
df = pd.DataFrame(valores)
df = df.drop('Boiling point',axis=1)
df = pd.get_dummies(df, drop_first=False)
```

Let's import the random forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
```

```python
rf = RandomForestRegressor()
```

```python
rf.get_params()
```

```python
params = {'max_depth':[2,4,6,8,10],'criterion':['squared_error','absolute_error'],'max_features':['sqrt','log2'],'min_impurity_decrease':[.09,.12,.25]}
```

```python
X = df.drop("Ionization Energies",axis=1).values
y = df["Ionization Energies"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
pip = Pipeline([("imputer",SimpleImputer()),("grdcv",GridSearchCV(rf, param_grid=params,cv=3))])
```

```python
pip.fit(X_train, y_train)
```

```python
pip.score(X_test, y_test)
```

```python
y_pred = pip.predict(X_test)
```

The random tree model had an improved score!, not surprising, given the flexibility of the random forests
